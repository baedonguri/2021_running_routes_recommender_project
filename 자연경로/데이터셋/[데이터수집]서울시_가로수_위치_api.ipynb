```python
import requests
import pandas as pd
import math
import json

gu_lst = ['중구', '용산구', '성동구', '광진구', '동대문구','중랑구','성북구',
          '강북구','도봉구','노원구','은평구','서대문구',
          '마포구','양천구','강서구','구로구','금천구',
          '영등포구','동작구','관악구','서초구','강남구','송파구','강동구'] # 종로구 데이터 null, 총 24개 구
key = '784d6464516568643839726d426b78'

gu_nm = [] #구명
ext_yno = [] #현존유무
width_nm = [] #가로명
wdpt_nm = [] #수목명
lng = [] #경도
lat = [] #위도


for gu in gu_lst:
    try:
        url = f'http://openapi.seoul.go.kr:8088/{key}/json/GeoInfoOfRoadsideTreeW/1/5/{gu}/'
        re = requests.get(url)
        json_result = json.loads(re.text)
        rjson = re.json()
        total_num = int(rjson['GeoInfoOfRoadsideTreeW']['list_total_count'])


        for i in range(1, math.ceil(total_num/1000)+1):
            end=i*1000
            start = end-1000 + 1

            if end > total_num:
                end = total_num

            url = f'http://openapi.seoul.go.kr:8088/{key}/json/GeoInfoOfRoadsideTreeW/{start}/{end}/{gu}/'
            re=requests.get(url)
            rjson=re.json()

            for u in rjson['GeoInfoOfRoadsideTreeW']['row']:
                gu_nm.append(u['GU_NM'])
                ext_yno.append(u['EXT_YNO'])
                width_nm.append(u['WIDTH_NM'])
                wdpt_nm.append(u['WDPT_NM'])
                lng.append(u['LNG'])
                lat.append(u['LAT'])
        print(gu, total_num,'건 저장 완료')
    except:
        print(gu, '저장 실패')
        pass
```

```python
df=pd.DataFrame({'구':gu_nm, '현존유무':ext_yno, '가로명':width_nm, '수목명':wdpt_nm, 'lon':lng, 'lat':lat})
df
```
