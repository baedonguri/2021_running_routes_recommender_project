```python
import pandas as pd
import numpy as np
```

```python
# 임의로 만든 이용내역(csv)파일
df = pd.read_csv('item.csv', encoding = 'utf-8')

# df1은 코스별 유사도를 판별하기위함
df1 = df.drop(columns = ['성별', '출발지'])

# df2는 가보지 않은 경로를 추천
df2 = df.drop(columns = ['성별', '출발지'])
df2.head()
```

```python
# 사용자별 경로 총 이용횟수 계산
df = df.fillna(0)
df['총 이용횟수'] = 0
for i in range(len(df)):
    df['총 이용횟수'][i] = df.loc[i][2] + df.loc[i][3] + df.loc[i][4] + df.loc[i][5] + \
                            df.loc[i][6] + df.loc[i][7]
df
```

```python
#결측값제거
df1 = df1.fillna(0)
df2 = df2.fillna(0)
```

```python
# index를 userid로 설정
df1 = df1.set_index('userid')
df2 = df2.set_index('userid')
df1.head()
```

```python
# cosine similarity 계산을 위한 transpose
df1_T = df1.transpose()
print(df1_T.shape)
df1_T.head()
```

### 사용자 이용내역 기반 코스와 코스간 유사도 판별

```python
from sklearn.metrics.pairwise import cosine_similarity
# 코사인 유사로를 이용해 반환된 넘파이 행렬을 코스로 매핑 후 dataframe으로 변환

# 전체 사용자 경로이용내역 기반 코스간 유사도를 판별
item_sim = cosine_similarity(df1_T, df1_T)

item_sim_df = pd.DataFrame(data = item_sim, index = df1.columns,
                          columns = df1.columns)
print(item_sim_df.shape)
item_sim_df
```

```python
# '명소구간경로' 와 다른 경로간 유사도를 오름차순 정렬
item_sim_df['명소구간경로'].sort_values(ascending = False)[:]
```

```python
# '자연자동경로' 와 다른 경로간 유사도를 오름차순 정렬
item_sim_df['자연자동경로'].sort_values(ascending = False)[:]
```

### 유저와 비슷한 러닝성향을 가진 다른 사용자 추천

```python
# 요소 1 -> 추천해줄 유저와 같은 경로를 뛰는 다른 사용자
# 'target' column은 유저가 제일 많이 이용한 경로를 찾아냄
df1['target'] = 0
for i in range(len(df1)):
    q = df1.loc[i+1]
    w = np.argmax(q)
    df1['target'][i+1] = q.index[w]
df1
```

```python
# 코스간 유사도 top 2를 기준으로 사용자를 1차 분리
q , w = item_sim_df[df1.loc[1][6]].sort_values(ascending = False)[:2].index
df1 = df1[(df1['target'] == q) | (df1['target'] == w)]
ind = df1.index
```

```python
# 요소 2 -> 성별이 같은 이들을 우선순위로 추천
# 요소 3 -> 총 이용횟수가 비슷한 이들을 우선순위로 추천
df_final = pd.concat([df1, df['성별'][ind], df['총 이용횟수'][ind]], axis = 1)
df_final
```

```python
for i in df_final.index:
    if i>1:
        if ((df_final['target'][1] == df_final['target'][i]) & (df_final['성별'][1] == df_final['성별'][i]) \
           & ((df_final['총 이용횟수'][1] - 10) < df_final['총 이용횟수'][i] < (df_final['총 이용횟수'][1] + 10))):
            print("1번 사용자와 가장 유사한 사용자의 정보")
            display(df_final.loc[i])
```

### 아이템 기반 인접이웃 협업필터링으로 개인화된 코스 추천

```python
def predict_course(df_arr, item_sim_arr):
    # 코사인 유사도와 코스간 이용횟수 내적후 평균
    pred = df_arr.dot(item_sim_arr) / np.array([np.abs(item_sim_arr).sum(axis = 1)])
    return pred
```

```python
df_pred = predict_course(df2.values, item_sim_df.values)
df_pred_matrix = pd.DataFrame(data = df_pred, index = df2.index,
                             columns = df2.columns)
print(df_pred_matrix.shape)
df_pred_matrix.head()
```

```python
from sklearn.metrics import mean_squared_error

# 가중치 부여뒤 예측 성능 평가 MearSquareError 계산

def get_mse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)

print("아이템 기반 모든 인접 이웃 MSE : ", get_mse(df_pred, df2.values))
```

```python
# 원하는 추천 갯수(n) 설정 뒤 예측 평점 계산
def predict_item_topsim(df_arr, item_sim_arr, n = 3):
    # 먼저 0으로 행렬 초기화
    pred = np.zeros(df_arr.shape)
    
    # 사용자 - 아이템 행렬의 열크기만큼 반복
    for col in range(df_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
        
        # 개인화된 예측값
        for row in range(df_arr.shape[0]):
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(df_arr[row, :][top_n_items].T)
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col, :][top_n_items]))
    return pred
```

```python
# 유사도를 기반으로 예측 MSE 계산
df_pred = predict_item_topsim(df2.values, item_sim_df.values, n = 3)
print("아이템 기반 인접 TOP - 3 이웃 MSE : ", get_mse(df_pred, df2.values))

# 계산된 예측 평점 데이터로 DataFrame 재생성
df_pred_matrix = pd.DataFrame(data = df_pred, index = df2.index, columns = df2.columns)
```

#### 사용자가 이용하지 않은 경로를 아이템 기반의 인접 이웃 협업필터링으로 추천

```python
def get_unused_route(df, userid):
    
    # 유저의 이용길을 Series로 변환
    user_route = df.loc[userid, :]
    
    # 0보다 크면 이미 이용한 경로
    already_used = user_route[user_route > 0].index.tolist()
    
    # 경로 list 변환
    route_list = df.columns.tolist()
    
    # 유저가 이미 이용한 길은 제외
    not_used_list = [route for route in route_list if route not in already_used]
    
    return not_used_list
```

```python
def recommend_route_by_userid(pred_df, userid, not_used_list, top_n = 2):
    # 예측값이 가장 높은 순으로 정렬
    recommend_route = pred_df.loc[userid, not_used_list].sort_values(ascending = False)[:top_n]
    return recommend_route

# 추천받을 사용자 입력
userid = 1
not_used_list = get_unused_route(df2, userid)

# 사용자가 이용하지 않은 경로
recommend_route = recommend_route_by_userid(df_pred_matrix, 1, not_used_list, top_n = 2)

# 아이템 기반의 인접 이웃 협업 필터링 알고리즘
recommend_route = pd.DataFrame(data = recommend_route.values, index = recommend_route.index,
                              columns = ['pred_score'])

print("1번 사용자에 대한 추천될 경로 예측 값")
print(recommend_route)
```
